```python
import sqlite3 as sql
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
conn = sql.connect('database.sqlite')
```

```python
pd.read_sql_query('select * from Salaries', conn)
```

```python
pd.read_sql_query('''PRAGMA table_info(Salaries)''', conn)
```

```python
#1 What is the average TotalPayBenefits for each JobTitle? exclude ('Not Provided', 'Not provided')
pd.read_sql_query('''select avg(TotalPayBenefits), JobTitle from Salaries
where EmployeeName Not In("Not Provided", "Not provided")
group by JobTitle
''', conn)
```

```python
#2 Which JobTitle has the most employees in each year? exclude ('Not Provided', 'Not provided')


pd.read_sql_query('''select count(employeename), JobTitle, year  from Salaries
where EmployeeName Not In("Not Provided", "Not provided")
group by  jobtitle
order by 1 desc
''', conn)
```

```python
#  3What is the total number of unique employees? exclude ('Not Provided', 'Not provided')


pd.read_sql_query('''select distinct count(employeename) from Salaries
where EmployeeName Not In("Not Provided", "Not provided")
''', conn)
```

```python
# 4 Show All Employees Ordered By Their TotalPayBenefits In Descending Order? exclude ('Not Provided', 'Not provided')

pd.read_sql_query('''select employeename, TotalPayBenefits from Salaries
where EmployeeName Not In("Not Provided", "Not provided")
order by TotalPayBenefits desc
''', conn)
```

```python
# 5 Show All Employees With A TotalPaybenefits Value Between 125,000 and 150,000 And A Job Title 'Firefighter'? exclude ('Not Provided', 'Not provided')
pd.read_sql_query('''select employeename, TotalPayBenefits, JobTitle from Salaries
where EmployeeName Not In("Not Provided", "Not provided") and jobtitle == "Firefighter" and TotalPayBenefits between 125000 and 150000
''', conn)
```

```python
#6 How many employees have no overtime pay or other pay? exclude ('Not Provided', 'Not provided')
pd.read_sql_query('''select count(employeename) from Salaries
where EmployeeName Not In("Not Provided", "Not provided") and OvertimePay == 0 and OtherPay == 0
''', conn)
```

```python
#7 Show top 10 employees having (overtimepay+OtherPay) in bar plot? exclude ('Not Provided', 'Not provided')
pd.read_sql_query('''select employeename, OvertimePay+OtherPay  "OT and OP" from Salaries
where EmployeeName Not In("Not Provided", "Not provided")
order by OvertimePay+OtherPay desc limit 10
''', conn)


```

```python
a = pd.read_sql_query('''select employeename, OvertimePay+OtherPay  "OT and OP" from Salaries
where EmployeeName Not In("Not Provided", "Not provided")
order by OvertimePay+OtherPay desc limit 10
''', conn)

a.plot(kind='bar', x='EmployeeName', y='OT and OP', color='green', figsize=(10,10))
plt.xlabel('EmployeeName')
plt.ylabel('OT and OP')
plt.title('Top 10 Employees')
plt.show()
```

```python
# 8. Show The Average of BasePay, OverTime, OtherPay for all employees in pie chart? exclude ('Not Provided', 'Not provided')
pd.read_sql_query('''select avg(BasePay), avg(OvertimePay), avg(OtherPay) from Salaries
where EmployeeName Not In("Not Provided", "Not provided")
''', conn)
```

```python
b = pd.read_sql_query('''select avg(BasePay), avg(OvertimePay), avg(OtherPay) from Salaries
where EmployeeName Not In("Not Provided", "Not provided")
''', conn)



b.plot(kind='pie', x='EmployeeName', y='OT and OP', color='green', figsize=(10,10))
plt.xlabel('EmployeeName')
plt.ylabel('OT and OP')
plt.title('Top 10 Employees')
plt.show()
```

```python
#9 Get the average TotalPayBenefits for each JobTitle, but only for those with an average pay greater than 75000? exclude ('Not Provided', 'Not provided')
pd.read_sql_query('''select avg(TotalPayBenefits), JobTitle from Salaries
where EmployeeName Not In("Not Provided", "Not provided")
group by JobTitle
having avg(TotalPayBenefits) > 75000
''', conn)
```

```python
#10 Get the number of employees (EmployeeName) for each JobTitle, where the count is greater than 10? exclude ('Not Provided', 'Not provided')
pd.read_sql_query('''select count(EmployeeName), JobTitle from Salaries
where EmployeeName Not In("Not Provided", "Not provided")
group by JobTitle
having count(EmployeeName) > 10
''', conn)
```
