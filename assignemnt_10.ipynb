```python
import sqlite3 as sql
import pandas as pd
```

```python
conn = sql.connect('factbook.db')
```

```python
def run_sql(q):
  return pd.read_sql(q, conn)
```

```python
q = '''select * from sqlite_master'''

run_sql(q)
```

```python
q = '''PRAGMA table_info(facts)'''

run_sql(q)
```

```python
q = '''select * from facts'''

run_sql(q)
```

```python
q = '''select * from cities'''

run_sql(q)
```

```python
###Q1 - 1. List all cities where the area is greater than 5000, along with their population.

q = '''select c.name, f.area  from cities c inner join facts f on c.facts_id = f.id
where f.area > 5000 '''


run_sql(q)
```

```python
####2. List the names of capital cities where the death rate is greater than 2% and the birth rate is less than 20%

q = '''select c.name Cities , f.birth_rate, f.death_rate from cities c
inner join facts f on f.id == c.facts_id
 where c.capital == 1 and f.death_rate > 2 and f.birth_rate < 20'''
run_sql(q)



```

```python
####3. List the cities with a population growth rate between 1% and 5%, sorted in desc by population growth.

q = '''select c.name Cities,f.population_growth from cities c
inner join facts f  on f.id == c.facts_id
where f.population_growth
between 1 and 5 order by f.population_growth desc'''
run_sql(q)
```

```python
####4. Find countries that start with "A" and do not have any cities listed in the cities table. The query should return the country name and city name. Using left join

q = '''select f.name Country ,c.name Cities from facts f
left join cities c on f.id = c.facts_id
where f.name like "A%" and c.facts_id is Null
'''

run_sql(q)




```

```python
####5. List all capital cities and their corresponding death rate using a LEFT JOIN.

q = '''select c.name Cities, f.death_rate Death_Rate from cities c 
        left join facts f 
        on 
        c.facts_id=f.id
        where c.capital ==1 '''

run_sql(q)




```
