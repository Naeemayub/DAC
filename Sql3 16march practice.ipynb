```python
import sqlite3 as sql
import pandas as pd
```

```python
conn = sql.connect('factbook.db')
```

```python
conn
```

```python
def read_sql(q):
    return pd.read_sql_query(q,conn)
```

```python
q = '''select * from sqlite_master'''

read_sql(q)
```

```python
q = '''pragma table_info(facts)'''
read_sql(q)
```

```python
q = '''pragma table_info(cities)'''
read_sql(q)
```

```python
q = '''select * from facts limit 20'''
read_sql(q)
```

```python
q = '''select * from cities limit 100'''
read_sql(q)
```

```python
q = '''select * from facts where name=='Pakistan' or name=='pakistan' '''
read_sql(q)
```

```python
q = '''select * from cities where facts_id == 132'''
read_sql(q)
```

```python
q = ''' select *
        from  facts f  
        inner join cities c 
        on f.id=c.facts_id 
    
    '''
read_sql(q)
```

```python
q = ''' select  c.name City,f.name Country, c.capital, c.population City_population
        from  facts f  
        inner join cities c 
        on f.id=c.facts_id 
    
    '''
read_sql(q)
```

```python
q = ''' select  c.name City,f.name Country, c.capital, c.population City_population
        from  cities c
        inner join facts f
        on f.id=c.facts_id 
    
    '''
read_sql(q)
```

```python
q = '''select c.name City, f.name Country from cities c inner join facts f on f.id=c.facts_id 
        where f.name =='Pakistan'
        '''
read_sql(q)
```

```python
q = '''select f.name Country, c.name city from facts f 
left join cities c 
on f.id=c.facts_id
where c.name is null
'''
read_sql(q)
```

```python
q = '''select f.name, c.name ,c.population from cities c 
    inner join facts f 
    on f.id = c.facts_id
    where c.population > 10000000'''
read_sql(q)
```
